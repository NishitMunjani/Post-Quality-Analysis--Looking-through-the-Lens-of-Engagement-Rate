```python
import plotly.express as px
import pandas as pd

# Data preparation
data = {
    'Category': ['Automobile', 'Beverages', 'Clothing', 'Electronics', 'Entertainment', 'Food', 'Jewellery', 'Makeup', 'Non-profit', 'Shoes'],
    'HF-HE': [0.27, 1.09, 0.20, 0.83, 0.14, 0.84, 0.25, 0.08, 0.03, 0.70],
    'HF-LE': [1.20, 1.63, 0.94, 0.61, 1.05, 1.01, 0.60, 0.73, 0.40, 0.37],
    'LF-HE': [0.96, 1.92, 0.46, 2.13, 0.56, 2.01, 0.96, 0.24, 0.67, 0.11],
    'LF-LE': [0.68, 0.36, 0.41, 0.07, 0.92, 0.60, 0.00, 0.25, 0.00, 0.91]
}

df = pd.DataFrame(data)

# Melt the dataframe to long format
df_melted = df.melt(id_vars='Category', var_name='Condition', value_name='Value')

# Create box plot
fig = px.box(df_melted, x='Condition', y='Value', color='Category', points="all", notched=True)

# Customize the layout
fig.update_layout(
    xaxis_title="Condition",
    yaxis_title="Value",
    title="Box Plot of Categories by Condition",
    template='plotly_white'
)

# Display the box plot
fig.show()
```
