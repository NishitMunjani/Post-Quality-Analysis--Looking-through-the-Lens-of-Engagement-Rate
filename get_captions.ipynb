```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import os
```

```python
post_df =  pd.read_excel(r"D:\Study\Project\Major project\Beverages\Final_engagement_Beverages.xlsx", sheet_name='LF-LE')
```

```python
post_df.head()
```

```python
# Step 1: Define the directory containing your JSON files
json_dir = 'D://Study//Project//Minor Project//Categories//Beverages//Brand Data'
# "D:\Study\Project\Minor Project\Categories\Automobile\Brand Data"  # Update this path accordingly

# Step 2: List all JSON files in the directory
json_files = [file for file in os.listdir(json_dir) if file.endswith('.json')]

# Step 3: Read each JSON file into a DataFrame and store them in a list
dfs = []
for file_name in json_files:
    file_path = os.path.join(json_dir, file_name)
    df = pd.read_json(file_path)
    dfs.append(df)

# Step 4: Concatenate or merge the DataFrames in the list into one DataFrame
brand_df = pd.concat(dfs, ignore_index=True)
```

```python
brand_df.head()
```

```python
merged_df = pd.merge(post_df, brand_df, left_on='Unnamed: 3', right_on='url', how='left')

# Update the 'Captions' column with the 'caption' values from the merged DataFrame
post_df['Captions'] = merged_df['caption']
```
