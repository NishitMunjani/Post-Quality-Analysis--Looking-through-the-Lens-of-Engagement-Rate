```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd

categories = [
    # 'Automobile',
    # 'Beverages',
    # 'Clothing',
    # 'Electronics',
    # 'Entertainment',
    # 'Food',
    # 'Jewellery',
    # 'Makeup',
    'Non-profit',
    'Shoes'
]

quadrant = [
    'HF-HE',
    'HF-LE',
    'LF-HE',
    'LF-LE'
]

for category in categories:
    for quad in quadrant:
        df = pd.read_excel(f"D://Study//Project//Major project//Quadrant Analysis//{category}//four_quadrant_influencers.xlsx", sheet_name = quad)
        main_df = pd.read_json(rf"D://Study//Project//Minor Project//Categories//Main_dataframe_{category}.json")
        df['Hashtags'] = ''

        for index, row in df.iterrows():
            if row['Urls'] in main_df['postUrl'].values:
                hashtags = main_df.loc[main_df['postUrl'] == row['Urls'], 'hashtags'].values[0]
                df.at[index, 'Hashtags'] = hashtags
        
        if 'Post_map' not in df.columns:
            df['Post_map'] = ''
        
        num_rows = len(df)
        filenames = [f"{i}.jpg" for i in range(num_rows)]
        df['Post_map'] = filenames
        df.to_excel(f'D:\Study\Project\Major project\Hashtag analysis\{category}\{quad}_hashtag\{quad}_hashtags.xlsx', index=False)
    
```
