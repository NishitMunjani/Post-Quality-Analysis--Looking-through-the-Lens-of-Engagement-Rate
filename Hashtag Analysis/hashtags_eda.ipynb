## Distribution of number of irrelevant hashtags for a post in each class

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Assuming 'new_df' is the combined DataFrame with all the categories and quadrants

# List of categories and quadrants
categories = ['Automobile', 'Beverages', 'Clothing', 'Electronics', 'Entertainment', 'Food', 'Jewellery', 'Makeup', 'Non-profit', 'Shoes']
quadrants = ['HF-HE', 'HF-LE', 'LF-HE', 'LF-LE']

# Iterate over each category
for category in categories:
    # Filter the DataFrame for the current category
    category_df = df[df['Category'] == category]

    # Create a figure with 4 subplots for each quadrant
    fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(20, 5))
    fig.suptitle(f'Category: {category}', fontsize=16)

    # Iterate over each quadrant and plot the histogram
    for idx, quadrant in enumerate(quadrants):
        quadrant_df = category_df[category_df['Quadrant'] == quadrant]
        
        # Drop NaN values
        quadrant_df = quadrant_df.dropna(subset=['num_irrelevant_hashtags'])
        
        # If the dataframe is not empty after dropping NaNs, plot the histogram
        if not quadrant_df.empty:
            ax[idx].hist(quadrant_df['num_irrelevant_hashtags'], bins=100)
            ax[idx].set_title(quadrant)
            ax[idx].set_xlabel('Number of Irrelevant Hashtags')
            ax[idx].set_ylabel('Frequency')
        else:
            ax[idx].set_title(quadrant)
            ax[idx].set_xlabel('Number of Irrelevant Hashtags')
            ax[idx].set_ylabel('Frequency')
            ax[idx].text(0.5, 0.5, 'No Data', horizontalalignment='center', verticalalignment='center', transform=ax[idx].transAxes)

    # Adjust layout and show the plot
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

## Average number of posts per hashtag for each distribution

```python
average_area_list = []
distribution_name = ['Affective', 'Behavioral', 'Cognitive', 'Negative']
for distribution in distribution_name:
    average_area_list.append(df[df['distribution']==distribution]['total_hashtag_popularity'].mean())

plt.bar(x=distribution_name, height=average_area_list)
plt.ylabel('Average number of posts for hashtags')
```

## Average post frequency for hashtags in each distribution

```python
average_area_list = []
distribution_name = ['Affective', 'Behavioral', 'Cognitive', 'Negative']
for distribution in distribution_name:
    average_area_list.append(df[df['distribution']==distribution]['total_hashtag_growth_rate'].mean())

plt.bar(x=distribution_name, height=average_area_list)
plt.ylabel('Average post frequency for hashtags')
```

## Area occupied by the main object in the post image in percentage
