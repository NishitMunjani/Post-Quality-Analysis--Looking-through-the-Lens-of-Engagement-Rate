```python
import pandas as pd
```

```python
categories = [
    'Automobile',
    'Beverages',
    'Clothing',
    'Electronics',
    'Entertainment',
    'Food',
    'Jewellery',
    'Makeup',
    'Non-profit',
    'Shoes'
]

quadrant = [
    'HF-HE',
    'HF-LE',
    'LF-HE',
    'LF-LE',
]

global_df = pd.DataFrame()
for category in categories:
    for quad in quadrant:
        path = f"D://Study//Project//Major project-1//Hashtag analysis//{category}//{quad}_hashtag//{quad}_hashtags.xlsx"
        df = pd.read_excel(path)
        df['Category'] = category
        df['Quadrant'] = quad
        global_df = pd.concat([global_df, df])
```

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# df['Average Words'] = pd.to_numeric(df['Average Words'], errors='coerce')

# Calculate the average follower/post for each quadrant within each category
contrast_colors = ['#90E0EF', '#00B4D8', '#0077B6', '#03045E']
cmap = LinearSegmentedColormap.from_list('contrast', contrast_colors, N=256)
average_df = new_df.groupby(['Category', 'Quadrant'])['irrelevant_hashtag_clustering_based'].mean().reset_index()

# Pivot the dataframe for better plotting
pivot_df = average_df.pivot(index='Category', columns='Quadrant', values='irrelevant_hashtag_clustering_based')

print('Average No. of Words across Quadrants and Categories')
print(pivot_df.values)
# Plotting
pivot_df.plot(kind='bar', figsize=(14, 8), colormap=cmap)
# plt.title('Average No. of Words across Quadrants and Categories')
plt.ylabel('Average No. of Words per Quadrants')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.legend(title='Quadrant')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
